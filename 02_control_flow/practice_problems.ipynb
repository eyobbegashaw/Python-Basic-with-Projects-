ðŸ”„ Problem 1: Number Pattern Generator

Problem Statement:

Create a program that generates the following number pattern:

 
1

22

333

4444

55555

```python
# Number Pattern Generator
print("Number Pattern:")

for i in range(1, 6):
    for j in range(i):
        print(i, end="")
    print()  # Move to next line

# Alternative solution using string multiplication
print("\nAlternative Solution:")
for i in range(1, 6):
    print(str(i) * i)
```

ðŸŽ¯ Problem 2: Student Grade Calculator

Problem Statement:

Create a program that:

Asks for the number of students

For each student, asks for name and 3 subject marks

Calculates total, average, and grade for each student

Displays a summary report

Solution:

```python
# Student Grade Calculator
print("=== Student Grade Calculator ===")

# Get number of students
num_students = int(input("Enter number of students: "))
students = []

# Collect data for each student
for i in range(num_students):
    print(f"\n--- Student {i+1} ---")
    name = input("Enter student name: ")
    
    # Get marks for 3 subjects
    marks = []
    for subject in ["Math", "Science", "English"]:
        mark = float(input(f"Enter {subject} marks (0-100): "))
        marks.append(mark)
    
    # Calculate total and average
    total_marks = sum(marks)
    average = total_marks / 3
    
    # Determine grade
    if average >= 90:
        grade = "A"
    elif average >= 80:
        grade = "B"
    elif average >= 70:
        grade = "C"
    elif average >= 60:
        grade = "D"
    else:
        grade = "F"
    
    # Store student data
    students.append({
        "name": name,
        "marks": marks,
        "total": total_marks,
        "average": average,
        "grade": grade
    })

# Display summary report
print("\n" + "="*50)
print("                 GRADE REPORT")
print("="*50)
print(f"{'Name':<15} {'Math':<8} {'Science':<8} {'English':<8} {'Total':<8} {'Average':<8} {'Grade':<6}")
print("-" * 50)

for student in students:
    marks = student["marks"]
    print(f"{student['name']:<15} {marks[0]:<8} {marks[1]:<8} {marks[2]:<8} {student['total']:<8} {student['average']:<8.1f} {student['grade']:<6}")

print("=" * 50)
```

ðŸª Problem 3: Inventory Management System

Problem Statement:

Create a simple inventory system that allows:

Adding products with name, price, and quantity

Viewing all products

Searching for a product

Calculating total inventory value

Exit option

Solution:

```python
# Simple Inventory Management System
print("=== Inventory Management System ===")

inventory = []

while True:
    print("\n--- Menu ---")
    print("1. Add Product")
    print("2. View All Products")
    print("3. Search Product")
    print("4. Calculate Total Value")
    print("5. Exit")
    
    choice = input("Enter your choice (1-5): ")
    
    if choice == '1':
        # Add Product
        print("\n--- Add New Product ---")
        name = input("Enter product name: ")
        price = float(input("Enter product price: $"))
        quantity = int(input("Enter product quantity: "))
        
        product = {
            "name": name,
            "price": price,
            "quantity": quantity,
            "value": price * quantity
        }
        inventory.append(product)
        print(f"âœ… Product '{name}' added successfully!")
    
    elif choice == '2':
        # View All Products
        print("\n--- All Products ---")
        if not inventory:
            print("No products in inventory.")
        else:
            print(f"{'No.':<4} {'Product Name':<15} {'Price':<10} {'Quantity':<10} {'Value':<10}")
            print("-" * 50)
            for i, product in enumerate(inventory, 1):
                print(f"{i:<4} {product['name']:<15} ${product['price']:<9.2f} {product['quantity']:<10} ${product['value']:<9.2f}")
    
    elif choice == '3':
        # Search Product
        print("\n--- Search Product ---")
        search_name = input("Enter product name to search: ").lower()
        found = False
        
        for product in inventory:
            if search_name in product['name'].lower():
                print(f"âœ… Found: {product['name']} - Price: ${product['price']:.2f} - Quantity: {product['quantity']} - Value: ${product['value']:.2f}")
                found = True
        
        if not found:
            print("âŒ Product not found in inventory.")
    
    elif choice == '4':
        # Calculate Total Value
        print("\n--- Inventory Value ---")
        if not inventory:
            print("No products in inventory.")
        else:
            total_value = 0
            total_quantity = 0
            
            for product in inventory:
                total_value += product['value']
                total_quantity += product['quantity']
            
            print(f"Total Products: {len(inventory)}")
            print(f"Total Quantity: {total_quantity}")
            print(f"Total Inventory Value: ${total_value:.2f}")
    
    elif choice == '5':
        # Exit
        print("Thank you for using Inventory Management System!")
        break
    
    else:
        print("âŒ Invalid choice! Please enter 1-5.")

# Sample usage demonstration
print("\n" + "="*50)
print("Sample Inventory Demo:")
print("="*50)

# Add some sample data
sample_products = [
    {"name": "Laptop", "price": 999.99, "quantity": 5},
    {"name": "Mouse", "price": 25.50, "quantity": 20},
    {"name": "Keyboard", "price": 75.00, "quantity": 15}
]

for product in sample_products:
    product["value"] = product["price"] * product["quantity"]
    inventory.append(product)

# Display sample inventory
print("Sample Inventory Created:")
print(f"{'No.':<4} {'Product Name':<15} {'Price':<10} {'Quantity':<10} {'Value':<10}")
print("-" * 50)
for i, product in enumerate(inventory, 1):
    print(f"{i:<4} {product['name']:<15} ${product['price']:<9.2f} {product['quantity']:<10} ${product['value']:<9.2f}")

# Calculate total
total_val = sum(product['value'] for product in inventory)
print(f"\nTotal Inventory Value: ${total_val:.2f}")
```
